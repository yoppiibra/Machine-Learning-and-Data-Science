```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots
```

```python
quarted = pd.read_csv(r'C:\Windows.old\Users\Data\anscombes_quartet\Anscombe_quartet_data.csv')
```

```python
quarted.head()
```

```python
# y 3.00 + 0.500 *
quarted['y_pred1'] = 3+0.5* quarted['x123']
quarted['residual1'] = quarted['y1'] - quarted['y_pred1']
```

```python
quarted.head()
```

```python
plt.style.use(['science','notebook','grid'])
```

```python
sns.scatterplot(x='x123', y='y1', data= quarted)
sns.lineplot(x='x123', y='y_pred1', data =quarted, color='red');
plt.vlines(quarted['x123'], quarted['y1'], quarted['y1']- quarted['residual1']);
```

```python
sns.kdeplot(quarted.residual1, fill =True);
```

```python
sns.scatterplot(x='y1', y='residual1', data=quarted)
plt.axhline(y=0, color='red', ls='--');
```
