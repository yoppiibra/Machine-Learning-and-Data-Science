 # Regularization
## Regularization attempts to minimize the RSS (residual sum of squares) and a penalty factor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import scienceplots
```

```python
df = pd.read_csv(r'C:\Users\Data\Advertising.csv\Advertising.csv', index_col=0)
```

```python
df.head()
```

```python
X = df.drop('Sales', axis=1)
y = df['Sales']
```

## Preprocessing  PolynomialFeatures

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomial_convert = PolynomialFeatures(degree=3, include_bias= False)
```

```python
polynomial_features = polynomial_convert.fit_transform(X)
```

## Split data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(polynomial_features, y, test_size=0.3, random_state=101)
```

## Scaling data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Ridge Regression (L2)

```python
from sklearn.linear_model import Ridge
```

```python
help(Ridge)
```

```python
ridge_model = Ridge(alpha=10)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
predict_ridge = ridge_model.predict(X_test)
```

## Uji metrics ridge

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
MAE = mean_absolute_error(y_test, predict_ridge)
MSE = mean_squared_error(y_test, predict_ridge)
RMSE = np.sqrt(MSE)
```

```python
MAE
```

```python
RMSE
```

```python
# perbandigan dengan train prediction
train_predict = ridge_model.predict(X_train)
MAE = mean_absolute_error(y_train, train_predict)
MAE
```

## Choosing alpha value with Cross Validation | K-fold

```python
from sklearn.linear_model import RidgeCV
```

```python
help(RidgeCV)
```

```python
RidgeCV_model = RidgeCV(alphas=(0.1, 1.0, 10), scoring='neg_mean_squared_error')
```

```python
RidgeCV_model.fit(X_train, y_train)
```

```python
RidgeCV_model.alpha_
```

```python
test_prediction = RidgeCV_model.predict(X_test)
```

## Uji metrics RidgeCV

```python
MAE = mean_absolute_error(y_test,test_prediction )
MSE = mean_squared_error(y_test, test_prediction)
RMSE = np.sqrt(MSE)
```

```python
MAE
```

```python
MSE
```

```python
RMSE
```

```python
# Training set performent
train_ridgecv = RidgeCV_model.predict(X_train)
MAE = mean_absolute_error(y_train,train_ridgecv)
MAE
```

```python
RidgeCV_model.coef_
```

# Lasso L2

```python
from sklearn.linear_model import Lasso, LassoCV
```

```python
help(LassoCV)
```

```python
lassoCV_model = LassoCV(eps = 0.1, n_alphas =100, cv=5)
```
